```python
%matplotlib inline
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
indiegogo = pd.read_csv('indiegogo_cleaned_dataset.tsv', sep='\t')
```

```python
indiegogo.drop('has_partner', axis=1).std(axis=0).to_frame(name="standard_deviation")
```

```python
from scipy.stats import norm
sns.distplot(indiegogo.title_len, fit=norm)
```

```python
sns.distplot(indiegogo.tagline_len, fit=norm)
```

```python
sns.distplot(indiegogo.pledges_count, fit=norm)
```

```python
indiegogo[indiegogo.pledges_count <= 1000].count().title_len
```

```python
indiegogo[indiegogo.pledges_count > 1000].count().title_len
```

```python
sns.distplot(indiegogo[indiegogo.pledges_count <= 1000].pledges_count, fit=norm)
```

```python
sns.distplot(indiegogo.collected_percentage, fit=norm)
```

```python
indiegogo[indiegogo.collected_percentage > 500].count().collected_percentage
```

```python
indiegogo[indiegogo.collected_percentage <= 500].count().collected_percentage
```

```python
sns.distplot(indiegogo[indiegogo.collected_percentage <= 500].collected_percentage, fit=norm)
```

```python
sns.distplot(indiegogo.target_amount, fit=norm)
```

```python
indiegogo.count().title_len
```

```python
indiegogo[indiegogo.target_amount > 100000].count().title_len
```

```python
sns.distplot(indiegogo[indiegogo.target_amount <= 15000].target_amount, fit=norm)
```

```python
barplot = sns.barplot(x='has_partner', y='pledges_count', data=indiegogo, estimator=np.std, palette="viridis")
barplot.set(xlabel='has_partner', ylabel='Standard Deviation pledges number')
```

```python
barplot = sns.barplot(x='has_partner', y='target_amount', data=indiegogo, estimator=np.std, palette="viridis")
barplot.set(xlabel='has_partner', ylabel='Standard Deviation amount required')
```

```python
barplot = sns.barplot(x='has_partner', y='collected_percentage', data=indiegogo, estimator=np.std, palette="viridis")
barplot.set(xlabel='has_partner', ylabel='Standard Deviation collected percentage')
```

```python
barplot = sns.barplot(x='has_partner', y='title_len', data=indiegogo, estimator=np.std, palette="viridis")
barplot.set(xlabel='has_partner', ylabel='Standard Deviation title length')
```

```python
barplot = sns.barplot(x='has_partner', y='tagline_len', data=indiegogo, estimator=np.std, palette="viridis")
barplot.set(xlabel='has_partner', ylabel='Standard Deviation tagline length')
```
