```python
%matplotlib inline
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
indiegogo = pd.read_csv('indiegogo_cleaned_dataset.tsv', sep='\t')
```

```python
indiegogo.std(axis=0).to_frame(name="standard_deviation")
```

```python
barplot = sns.barplot(x='has_partner', y='pledges_count', data=indiegogo, estimator=np.std, palette="viridis")
barplot.set(xlabel='has_partner', ylabel='Standard Deviation pledges number')
```

```python
barplot = sns.barplot(x='has_partner', y='balance', data=indiegogo, estimator=np.std, palette="viridis")
barplot.set(xlabel='has_partner', ylabel='Standard Deviation amount required')
```

```python
barplot = sns.barplot(x='has_partner', y='collected_percentage', data=indiegogo, estimator=np.std, palette="viridis")
barplot.set(xlabel='has_partner', ylabel='Standard Deviation collected percentage')
```

```python
barplot = sns.barplot(x='has_partner', y='title_len', data=indiegogo, estimator=np.std, palette="viridis")
barplot.set(xlabel='has_partner', ylabel='Standard Deviation title length')
```

```python
barplot = sns.barplot(x='has_partner', y='tagline_len', data=indiegogo, estimator=np.std, palette="viridis")
barplot.set(xlabel='has_partner', ylabel='Standard Deviation tagline length')
```
