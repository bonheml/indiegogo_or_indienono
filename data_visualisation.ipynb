```python
%matplotlib inline
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
from copy import deepcopy
```

```python
indiegogo = pd.read_csv('indiegogo_cleaned_dataset_500.tsv', sep='\t')
```

```python
indiegogo[indiegogo.collected_percentage > 500].category_slug.value_counts()
```

```python
with sns.plotting_context("notebook"):
    plt.figure(figsize=(20, 20))
    sns.jointplot(x="pledges_count", y="collected_percentage", data=indiegogo, kind="reg")
```

```python
with sns.plotting_context("notebook"):
    plt.figure(figsize=(20, 20))
    sns.jointplot(x="balance", y="collected_percentage", data=indiegogo, kind="reg")
```

```python
with sns.plotting_context("notebook"):
    plt.figure(figsize=(20, 20))
    sns.jointplot(x="pledges_count", y="balance", data=indiegogo, kind="reg")
```

```python
with sns.plotting_context("notebook", font_scale=1.5):
    plt.figure(figsize=(20, 20))
    sns.countplot(y="category_slug", data=indiegogo)
```

```python
with sns.plotting_context("notebook"):
    plt.figure(figsize=(20, 20))
    g = sns.PairGrid(indiegogo,
                 x_vars=["currency_code", "has_partner"],
                 y_vars=["title_len", "tagline_len", "collected_percentage", "pledges_count", "balance"])
    g.map(sns.barplot, palette="viridis")
```

```python
dance = indiegogo[indiegogo.category_slug == "dance"]
with sns.plotting_context("notebook", font_scale=1.5):
    plt.figure(figsize=(20, 20))
    ax = sns.kdeplot(dance.collected_percentage, dance.balance, cmap="Reds", shade=True, shade_lowest=False)
```

```python
indiegogo_wo = indiegogo[indiegogo.balance < 10000]
indiegogo_wo = indiegogo_wo[indiegogo_wo.collected_percentage < 500]
```

```python
with sns.plotting_context("notebook", font_scale=1.5):
    plt.figure(figsize=(20, 20))
    ax = sns.kdeplot(indiegogo_wo.collected_percentage, indiegogo_wo.balance, cmap="Reds", shade=True, shade_lowest=False)
```

```python
with sns.plotting_context("notebook", font_scale=1.5):
    plt.figure(figsize=(20, 20))
    ax = sns.kdeplot(indiegogo_wo.pledges_count, indiegogo_wo.balance, cmap="Reds", shade=True, shade_lowest=False)
```
